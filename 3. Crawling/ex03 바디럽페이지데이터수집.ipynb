##### 데이터 전처리
- 앞, 뒤 공백을 제거
- 순수한 리뷰 데이터만 출력(네이버 페이 구매평이라는 글자 제거)

```python
# 앞, 뒤 공백을 없애는 메소드
review[0].text.strip()
```

```python
temp = review[0].text.strip()
# 'X' in string -> string 안에 'X'라는 글자가 있다면 True, 없다면 False
'네이버 페이 구매평' in temp
```

```python
# 슬라이싱을 통해 접근
# 리뷰 내용이 매번 달라지므로 앞에서부터 접근 X, 뒤에서부터 접근!
if '네이버 페이 구매평' in temp :
    temp = temp.replace(temp[-38:],'')
print(temp)
```

```python
reviewList = [];
for i in review :
    if '네이버 페이 구매평' in i.text : 
        reviewList.append(i.text.strip().replace(i.text.strip()[-38:],''))
    else :
        reviewList.append(i.text.strip())
```
