### requests 라이브러리 사용법

```python
# 라이브러리 불러오기
import requests as req
# 서버에게 데이터를 요청 및 응답받기 위한 라이브러리
```

```python
res=req.get('https://www.naver.com/')
# Response[200]번대 : 성공적으로 요청 및 응답이 이루어졌다는 의미
# Response[400]번대 : 클라이언트(요청) 부분에서 오류가 발생했다는 의미
# Response[500]번대 : 서버(응답) 부분에서 오류가 발생했다는 의미
```

```python
res.text
```

### BeautifulSoup 라이브러리 불러오기

```python
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해할 수 있는 파이썬 객체 형태로 변환해주는 라이브러리
```

```python
# bs(parsing할 html문서, 'parsing 방법')
# lxml : 빠르고 관대함
soup = bs(res.text, 'lxml')
soup
```

```python
res =req.get('https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EA%B5%B0%EC%88%9C&oquery=%E3%85%87&tqi=iLtfflprvxZss7gw8YCssssstlC-458344')
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
# 태그의 위치를 찾응 방법
# 1. 내가 가져오고 싶은 contents 위에 마우스 우클릭 → 검사
# 2. 개발자모드 좌측 상단 커서 모양 아이콘 클릭 → 가져오고 싶은 contents 클릭
```

##### 네이버 페이지에서 정보를 가져와보자
- 뉴스, 인플루언서, 통합 등과 같은 글자를 수집해보자

```python
# a 태그를 가져오는 방법
soup.select('a')
# 페이지에 있는 모든 a태그가 가져와졌다
```

```python
# a태그 중에 클래스가 tab인 애들만 데려와보자
naver_tab = soup.select('a.tab')
# 여기중에 '통합'이라는 contents를 가져와보자
```

```python
naver_tab[0]
```

```python
soup.select_one('a.tab')
# select_one : 여러가지 요소 중에서 첫번째 요소만 단수형으로 가져와준다
# select : 여러가지 요소를 복수형(list)으로 가져와준다
```

```python
naver_tab[0].text
```
