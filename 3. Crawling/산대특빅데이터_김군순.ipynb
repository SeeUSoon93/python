### Question 3
- 문제2

```python
# 웹브라우저를 제어
from selenium import webdriver as wb
# 웹에게 값을 입력하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys
# html문서에서 태그와 선택자의 위치를 찾기 위한 라이브러리
from selenium.webdriver.common.by import By
# 컴퓨터가 이해하기 쉬운 객체로 변환
from bs4 import BeautifulSoup as bs
# DataFrame
import pandas as pd
# 시간제어
import time
```

```python
driver = wb.Chrome()

product_name = []
product_price = []
product_category = []
rank = []

for i in range(0,20) :
    driver.get('http://corners.gmarket.co.kr/Bestsellers')
    link = driver.find_elements(By.CLASS_NAME,'image__lazy.image__load')
    link[i].click()
    soup = bs(driver.page_source,'lxml')
    
    name = soup.select_one('h1.itemtit')
    price = soup.select_one('strong.price_real')    
    category = soup.select_one('body>div.location-navi>ul>li:nth-child(3)>a')
        
    product_name.append(name.text)
    product_price.append(price.text)
    product_category.append(category.text)
    rank.append(i+1)
    time.sleep(2)
        
driver.close()

dic = {'상품 이름':product_name, '가격':product_price, '카테고리': product_category}
re_df = pd.DataFrame(dic,index = rank)
re_df
```

```python
import requests as req # 서버와의 요청 및 응답을 위한 라이브러리
from bs4 import BeautifulSoup as bs # 컴퓨터가 이해하기 쉬운 객체로 변환하는 라이브러리
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/index.htm'
h = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
res = req.get(url,headers = h)
soup = bs(res.text, 'lxml')

title = soup.select('div.ellipsis.rank01 a')
singer = soup.select('div.ellipsis.rank02 a+span')

title_list = []
singer_list = []
index =[]
for i in range(len(singer)) :
    title_list.append(title[i].text)
    singer_list.append(singer[i].text)
    index.append(i+1)

dic = {'곡명':title_list, '가수':singer_list}
rank = pd.DataFrame(dic,index = index)

rank.to_csv('멜론TOP100.csv', encoding='euc-kr')

rank
```

```python
# 웹브라우저를 제어
from selenium import webdriver as wb
# 웹에게 값을 입력하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys
# html문서에서 태그와 선택자의 위치를 찾기 위한 라이브러리
from selenium.webdriver.common.by import By
# 컴퓨터가 이해하기 시ㅜ운 객체로 변환
from bs4 import BeautifulSoup as bs
# DataFrame
import pandas as pd
# 시간제어
import time
```

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
btn = driver.find_element(By.CLASS_NAME,'c_05')
menu_list = []
price_list = []
time.sleep(2)

try :
    while(True) :        
        btn.click()
        time.sleep(2)
except :
    print('종료')
    
soup = bs(driver.page_source,'lxml')
menu = soup.select('h4.h.fz_03')
price = soup.select('div.item-price>strong')

for i in menu :
    if '[8월' in i.text : 
        menu_list.append(i.text.strip().replace(i.text.strip()[:8],''))
    else :
        menu_list.append(i.text.strip())
        
for i in price :
    price_list.append(i.text.strip()+'원')
    
driver.close()
    
dic = {'메뉴':menu_list, '가격':price_list}
re_df = pd.DataFrame(dic)
re_df.to_csv('한솥menu.csv', encoding='euc-kr')
re_df
```
