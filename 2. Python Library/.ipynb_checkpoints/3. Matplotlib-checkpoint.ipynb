## matplotlib
- 그래프로 시각화가 가능한 라이브러리
- 파이썬에서 데이터를 차트나 플롯으로 그려줌
- 차트의 종류는 산점도(scatter), 막대(bar), 원형(pie), 꺾은 선, 히스토그램 등 다양함

```python
import matplotlib.pyplot as plt
```

### 기본적인 그래프 그려보기

```python
# Line plot (선 그래프)
y = [2, 12, 6]

# plot(y요소)
# 데이터를 하나만 넣으면 y축으로 간다. 이때 x축은 인덱스 0부터 자동부여 
plt.plot(y)

# show() → 차트를 표현하는 print()
plt.show()
```

```python
# line chart (2ea 데이터)
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, ls = '-.', c = '#97903C', marker='|')
```

- 라인 스타일(Line Stlye)
|Chatacter|Description|
|:-------:|:---------:|
|'-'|Soild line stlye|
|'--'|Dashed line stlye|
|'-.'|Dash-dot line stlye|
|':'|Dotted line stlye|

- 라인 컬러(Line color)
|Chatacter|color|Chatacter|color|
|:-------:|:---------:|:-------:|:---------:|
|'b'|Blue|'m'|Magenta|
|'g'|Green|'y'|yellow|
|'r'|Red|'k'|black|
|'c'|cyan|'w'|white|

- 스타일 옵션 종류
|스타일옵션종류|마커 종류정의|약자|
|:-------:|:---------:|:-------:|
|color|선 색깔|c|
|linewidth|선 굵기|lw|
|linestyle|선 스타일|ls|
|marker|마커 종류||
|markersize|마커 크기|ms|
|markeredgecolor|마커 선 색깔|mec|
|markeredgewidth|마커 선 굵기|mew|
|markerfacecolor|마커 내부 색깔|mfc|

- maker 종류
|Chatacter|Description|Chatacter|Description|
|:-------:|:---------:|:-------:|:---------:|
|'.'|Point marker|'1'|Tri_down marker|
|'o'|Circle marker|'2'|Tri_up marker|
|'v'|Triangle_down marker|'3'|Tri_left marker|
|'^'|Triangle_up marker|'4'|Tri_right marker|
|'<'|Triangle_left marker|'\*'|Star marker|
|'>'|Triangle_right marker|'h','H'|Hexagon1,2 marker|
|'s'|Square marker|'+'|Plus marker|
|'p'|Pentagon marker|'D'|Diamond marker|
|'\|'|Vline marker|'-'|Hline marker|

```python
# 직접 그려보자
import numpy as np

x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

# $특정문자$ → 실제로 넣고 싶은 문자 기호
# color → 헥스코드로 가능

plt.plot(x,y, ls=':', color='b', marker='d', mec ='g', mfc='b', ms='10' )

# 그래프에 표시도니 숫자의 범위지정
# 범위를 지정해서 그래프 확대나 축소 출력가능
plt.xlim(-1,8)
plt.ylim(0,9)

# 눈금을 지정
plt.xticks(np.arange(0, 8, 2))
plt.yticks(np.arange(0, 8, 2))

# grid() :격자무늬 생성
plt.grid()

plt.show()
```

### 여러 개의 꺾은 선차트 하나의 공간에 그리기

```python
x = [2,3,4,5,6,7,8]
y1 = [5,7,3,9,1,8,6]
y2 = [9,5,7,2,3,7,1]

plt.xlabel('x')
plt.ylabel('y', rotation=0)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# 한글이 깨진다 → matplotlib 한글이 깨졌을 때 사용
from matplotlib import rc # 폰트를 설정하는 모듈
rc('font', family='SUITE')
# font : 폰트를 설정하겠다.
# family : 글꼴집합이라는 의미로 글꼴명을 적어주면 됨!
```

```python
# 여러개의 차트 하나의 공간에 그려보기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label='x데이터')
plt.plot(y, label='y데이터')
plt.plot(z, label='z데이터')

plt.legend()

# chart 제목 부여
plt.title('제목', loc= 'right', size= '14')

plt.show()
```

## 요일별로 사고건수를 count 해보자!
- 요일별로 사고건수를 확인하되, 요일 순서대로(월~일) 출력

```python
# 요일별 발생한 사고건수를 카운트
y = data['요일'].value_counts()
y
```

```python
# 정렬을 하더라도 가나다순으로 되어버리지 요일별 정렬이 불가능함
y.sort_index()
```

```python
# 재정렬 하는 방법
y = y[['월','화','수','목','금','토','일']]
y
```

## Bar 차트

```python
x = y.index
plt.bar(x,y, color = 'lightblue')

plt.ylim(450,650)

plt.xlabel('요일', labelpad=20, fontdict={'weight': 'bold', 'size':12})
plt.ylabel('사고건 수', labelpad=20, fontdict={'weight': 'bold', 'size':12})
plt.title('요일별 교통사고 건수',pad=40, fontdict={'weight': 'bold', 'size':14})

# text(x좌표, y좌표, 넣고자 하는 값)
for i in range(len(y)) :
    plt.text(x[i],y[i],f'{y[i]}')

plt.show()
```

## '차대차'사건 중 사상자 수가 많은 '발생지 시도'를 알아보고 시각화
- 키워드 : 차대차, 사상자 수, 발생지 시도

```python
# 먼저 사고유형_대분류가 차대차인 행들만 출력
car = data[data['사고유형_대분류'] == '차대차']
```

```python
# groupby(컬럼명).단일함수()
chart_car = car[['발생지시도','사상자수']].groupby(['발생지시도']).sum()
```

```python
# line 그리기
x = chart_car.index
y = chart_car['사상자수']
```

```python
plt.figure(figsize=(15,8)) # figure(figsize=(x,y)) → chart의 사이즈 지정. 항상 차트 위에 생성
plt.plot(x,y)

plt.xlabel('발생지 시도', labelpad=20, fontdict={'weight': 'bold', 'size':15})
plt.ylabel('사상자 수', labelpad=20, fontdict={'weight': 'bold', 'size':15})

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.bar(x,y)

plt.xlabel('발생지 시도', labelpad=20, fontdict={'weight': 'bold', 'size':15})
plt.ylabel('사상자 수', labelpad=20, fontdict={'weight': 'bold', 'size':15})

plt.show()
```
