#### Question3
아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오. 

```python
list1 = [1,50,27,19,37,33,5,3,48]

odd_cnt = 0
even_cnt = 0

for i in list1 :
    if i%2 == 0 :
        even_cnt += 1
    else :
        odd_cnt += 1
        
print(f"짝수 {even_cnt}개, 홀수 {odd_cnt}개")
```

#### Question4
다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 프로그램을 작성하시오. (단, 학생의 숫자가 늘어나도 동작할 수 있도록 고려하시오.)

```python
score_list = [['김미희',89,78,99],
             ['서대희',67,88,77],
             ['강예진',100,55,96]]
avg = []
grade = []
print1 = ['이름','Python','Java','DB','평균','학점']
print2 = []

for i in score_list :
    sum = 0;
    for k in range(1,len(i)) :
        sum += i[k]        
    avg.append(round(sum/(len(i)-1)))

for i in avg :
    if i >= 90 :
        grade.append('A')
    elif i >= 80 :
        grade.append('B')
    elif i >= 70 :
        grade.append('C')
    elif i >= 60 :
        grade.append('D')
    else :
        grade.append('F')
for i in range(len(score_list)) :    
    print2.append(score_list[i])
    score_list[i].append(avg[i])
    score_list[i].append(grade[i])

print3 = ['평균']
for i in range(1,len(score_list)+1) :
    sum =0;
    for k in score_list :         
        sum += k[i]        
    print3.append(round(sum/len(score_list)))

print1 = [print1] +print2 +[print3]

for i in print1 :
    print(i,end='\t\n')
```

```python
X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]

import matplotlib.pyplot as plt

plt.grid()
plt.xlabel('X')
plt.ylabel('Y', rotation=0)

plt.plot(X, Y1, c='r')
plt.plot(X, Y2, c='b')
plt.legend(['Y1','Y2'])
plt.show()
```

#### Question6
Traffic_Accident_2017.csv 데이터를 이용하여 ‘차대 사람’인 사건 중 야간에 발생한 사건에 대해 아래 주어진 조건을 지켜 다음과 같이 시각화 하시오.

※ 조건 내용 반드시 확인할 것  
 조건 1. 사고유형_대분류가 ‘차대사람’ 이면서 발생시간대가 ‘야간’인 경우의 사고 건수 카운트  
 조건 2. y축의 범위는 0~200으로 설정할 것  
 조건 3. 차트 내부 글씨는 한글로 표시할 것  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

data = pd.read_csv('./Traffic_Accident_2017.csv', encoding='euc-kr')
car_people = data[data['사고유형_대분류']=='차대사람']
night = car_people[car_people['주야']=='야간']
day = night['요일'].value_counts()
day = day[['월', '화', '수', '목', '금', '토', '일']]

x = day.index
plt.bar(x,day)
plt.ylim(0,200)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대사람 교통사고 건수')

plt.show()
```
