```python
import matplotlib.pyplot as plt

font1 = {'color': 'red',
         'weight': 'bold',
         'size': 14
         }
font2 = {'color': 'blue',
         'weight': 'normal',
         'size': 'x-large'
         }

plt.plot([1, 2, 3, 4], [1, 10, 5, 2])
plt.xlabel('김군순의 외모', labelpad=20, fontdict=font1, loc='right')
plt.ylabel('김군순의 선성향', labelpad=20, fontdict=font2, loc='top')
plt.show()
```
