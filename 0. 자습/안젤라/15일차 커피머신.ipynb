### 커피머신

1. 처음에 모든 자료를 보여줌
- 물/우유/커피 남은 양 및 들어있는 돈

2. 메뉴를 주문했을 때 재료가 부족하면 재료가 부족하다는 문구

3. 동전받기
- 50원, 100원, 500원짜리 받기
- 돈 부족하게 받으면 환불하고 커피 안줌
- 돈 넘게 받으면 거스름돈 줌

```python
import time

menu = {"에스프레소":
        {"재료":
         {"물": 50, "커피": 18},
         "가격": 150
        },
        "카페라떼":
        {"재료":
         {"물":200, "우유":150, "커피": 24},
         "가격": 350
        },
        "카푸치노":
        {"재료":
         {"물":250, "우유":100, "커피": 24},
         "가격": 500
        }
       }
money = 0
resources = {
    "물": 3000,
    "우유": 2000,
    "커피": 1000,
}
```

```python
# 남은 재료 알려주는 함수
def report(resources) :
    print(f"""====남은 재료====    
물: {resources['물']}ml
우유: {resources['우유']}ml
커피: {resources['커피']}g

잔돈: {money}원
==============
""") 

# 메뉴 알려주는 함수


def menu_list(menu) :
    print(f"""=====메뉴=====    
에스프레소 : {menu['에스프레소']['가격']}원
카페라떼 : {menu['카페라떼']['가격']}원
카푸치노 : {menu['카푸치노']['가격']}원
=============
""")
    
# 돈 받는 함수
def insert_coin() :
    
    print("====동전을 넣어주세요====")
    coin = int(input("50원짜리 개수 : "))*50
    coin += int(input("100원짜리 개수 : "))*100
    coin += int(input("500원짜리 개수 : "))*500
    print()
    return coin

# 재료 남았는지 판단하는 함수
def isResourceOk(order_resources):
    for item in order_resources:
        if order_resources[item] > resources[item]:
            time.sleep(1)
            print(f"죄송합니다. {item}의 양이 부족합니다.")
            return False
    return True

# 돈 맞게 넣었는지 판단하는 함수
def success(money_received, drink_cost):    
    if money_received >= drink_cost:
        change = money_received - drink_cost
        time.sleep(1)
        print()
        print(f"잔돈 {change} 원입니다.")
        global money
        money += drink_cost
        return True
    else:
        time.sleep(1)
        print()
        print("돈이 부족합니다. 다시 넣어주세요")
        return False

# 커피 만드는 함수
def make_coffee(drink_name, order_resources):
    for item in order_resources:
        resources[item] -= order_resources[item]
        print()
        time.sleep(1)
    print(f"주문하신 {drink_name} 나왔습니다!")
    print()
```

```python
while True :
    report(resources)
    menu_list(menu)
    order = input("무엇을 주문하시겠습니까? ")
    print()
    if order == "종료" :
        print()
        print("다음에 또 이용해 주세요")
        break;
    else :
        drink = menu[order]
        if isResourceOk(drink["재료"]):
            pay = insert_coin()
            if success(pay,drink["가격"]):
                make_coffee(order,drink["재료"])
```
