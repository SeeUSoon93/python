### 문자열 포맷코드
- 문자열 내부에 값 삽입


```python
day = 24
month = 7

s = '오늘은 %d월 %d일 입니다.' %(month, day)
print(s)
```

```python
# format 함수

s = "오늘은 {}월 {}일 입니다.".format(month,day)
s
```

```python
# f 문자열 

s = f"오늘은 {month}월 {day}일 입니다."
s
```

```python
x = 100
y = 200
sum2 = x+y
print( f"{x}와 {y}의 합은 {sum2}입니다.")
print("%d와 %d의 합은 %d입니다."%(x,y,sum2))
print( "{}와 {}의 합은 {}입니다.".format(x,y,sum2))
```
