### 문자열 인덱싱과 슬라이싱
- 인덱싱 : 요소 하나를 가리키는 것 → 하나하나를 "가리킨다"
- 스라이싱 : 요소들을 잘라 놓오는 것 → 범위를 지정해서, 어디부터 어디까지를 잘라낸다

    * 인덱스 번호는 앞에서 0부터, 뒤로 세면 -1부터 시작
    * 슬라이스 - [포함:미포함]

```python
s = 'My name is GS'
```

```python
print(s[11])
```

```python
print(s[-2],s[-1])
```

```python
print(s[0:2])
print(s[3:7])
print(s[8:10])
print(s[11:13])
```
